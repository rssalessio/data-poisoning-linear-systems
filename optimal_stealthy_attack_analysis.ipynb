```python
A =data[2]['AB_unpoisoned'][:, :-1].detach().numpy()


eigenvalues = [np.linalg.eigvals(A)]

iterations = len(data[2]['AB_poisoned'])

for i in range(0, iterations):
    A_pois = data[2]['AB_poisoned'][i][:, :-1]
    eig_A_pois = np.linalg.eigvals(A_pois)
    eigenvalues.append(eig_A_pois)

eigenvalues = np.array(eigenvalues)
```

```python
r = np.abs(eigenvalues[:])
theta = np.angle(eigenvalues[:])

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rmax(1.05)
ax.set_rmin(0.7)
ax.set_rticks([0.85, 1])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```
